```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/penguins_size.csv")
```

```python
detailOfPenguin = ["species","island","culmen_length_mm", "culmen_depth_mm", "flipper_length_mm", "body_mass_g", "sex"]
detailOfPenguin2 = ["culmen_length_mm", "culmen_depth_mm", "flipper_length_mm", "body_mass_g"]

df = df.dropna(subset=detailOfPenguin)
df = df[detailOfPenguin].copy()
# df1 = df[detailOfPenguin2].copy()
```

```python
km = KMeans()
```

```python
# df1['cluster'] = km.fit_predict(df1)
# df1['island'] = df['island']
# df1['sex'] = df['sex']
# df1['']

df['cluster'] = km.fit_predict(df[detailOfPenguin2])
```

```python
df[['cluster', 'island', 'species', 'sex']].value_counts(sort=False)
```

```python
df.describe()
```

```python
X = df.iloc[:, [2, 4]].values
inert_list = []

X[1:10]
```

```python
for i in range(1,20):
  kmeans = KMeans(n_clusters=i, random_state= 100)
  kmeans.fit(X);
  inert_list.append(kmeans.inertia_)
```

```python
inert_list[1:10]
```

```python
plt.figure(figsize=(10, 5))
plt.plot(range(1,20), inert_list)
plt.xlim(1,8)
plt.show()
```

```python
kmeans = KMeans(n_clusters= 6, random_state= 101)

kmeans.fit(X)
```

```python
kmeans.__dict__
```

```python
plt.figure(figsize= (10, 7))
plt.scatter(X[:, 0], X[:, 1], c = kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=500, color = 'r', marker='*')
```
